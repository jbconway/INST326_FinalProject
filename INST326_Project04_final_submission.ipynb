```python
# imports
import tkinter as tk # import the tkinter library
from tkinter import ttk # import the ttk module for widgets
from tkinter import filedialog # import the filedialog module for file selection
import datetime # module for working with dates and times
import json # import module for handeling JSON files
import notes_module as notesModule # custom module for handeling notes
from tkinter import messagebox # import message box for error handeling 

# The MainWindow class serves as the main application window for the notebook GUI application, 
# providing a user interface to manage notes through a variety of widgets and controls.
# The class initializes the main window, sets up the main frame, and creates buttons for common 
# operations such as creating a new note, opening a notebook, and saving the notebook.
# Additionally, the class contains methods to handle user interactions, such as opening a new note 
# form, clearing the notes display, refreshing the notes view, and saving or opening notebook files.

class MainWindow(tk.Tk):
    def __init__(self):  #initialize the main window
        super().__init__() # initialize it as a tkinter window
        
        self.geometry("700x500") # set the default window size # improvement : made window larger
        self.title('Notebook') #set the default window title
        self.config(bg='lightpink') # background color of the window
        self.notebook = [] # initialize an attribute named 'notebook' as an empty list
        self.notes = [] # list of note objects
        self.note = {}
        # Create a new note object by calling the Note class from the notesModule module.
        self.note = notesModule.Note()  

        # creates a main frame covering the entire window 
        self.frame_main = tk.Frame(self)
        self.frame_main.grid(row=0, column=0, sticky="nsew")  # Use 'grid' instead of 'pack'
        self.grid_rowconfigure(0,weight = 1) # allow resizing in row 0
        self.grid_columnconfigure(0,weight = 1) # allow resizing in column 0
        self.frame_main.config(bg='light pink') # set background to pink

        # Create the canvas and scrollbar
        self.scroll_bar = tk.Scrollbar(self.frame_main, orient=tk.VERTICAL)
        self.canvas_notes = tk.Canvas(self.frame_main, yscrollcommand=self.scroll_bar.set)

        # Place the canvas and scrollbar in the grid
        self.canvas_notes.grid(row=0, column=1,columnspan=4 ,padx=(30,0), pady = 20, sticky=tk.N+tk.S+tk.W)
        self.scroll_bar.grid(row=0, column=5,pady = 20, sticky=tk.N+tk.S+tk.W)
        
        # Link the canvas's y-scroll command with the scrollbar's set method
        self.scroll_bar.config(command=self.canvas_notes.yview)

        # Create a frame for notes and place it inside the canvas
        self.frame_notes = tk.Frame(self.canvas_notes, bg='gray')
        self.canvas_notes.create_window((0, 0), window=self.frame_notes, anchor='nw')

        # Bind the configure event to update the scroll region of the canvas
        self.frame_notes.bind("<Configure>", self.update_scroll_region)

        # Configure weights to allow resizing of the main frame
        self.frame_notes.columnconfigure(0, weight=1)
        self.frame_notes.rowconfigure(0, weight=1)
        self.frame_notes.rowconfigure(1, weight=0)


        #define some buttons and put them in a grid in frame_main
        # create new note button - opens a new note form
        btn01=tk.Button(self.frame_main, text='Create New Note', command=self.new_note)
        btn01.grid(padx=5, pady=2, row=2, column=2)
        
        # open notebook button - opens a notebook and displays its notes in the window
        btn02=tk.Button(self.frame_main, text='Open Notebook', command=self.open_notebook)
        btn02.grid(padx=5, pady=2, row=2, column=3)
        
        # save notebook button - saves the notebook and refreshes the notes display
        btn03=tk.Button(self.frame_main, text='Save Notebook\nand Refresh',  command=lambda: self.save_notebook(self.notebook))
        btn03.grid(padx=5, pady=2, row=2, column=4) 
        
        # quit button
        btn04=tk.Button(self.frame_main, text='Quit', command=self.destroy)
        btn04.grid(padx=5, pady=2, row=2, column=5)

    
    # new_note opens a new toplevel window with a note form.
    # when the new note's submit button is pressed, the new note is added to the self.notebook
    def new_note(self): 
        note_form = NoteForm(self, self.notebook, self.notes) # creates a new note form
        return None

    # method for clearing old content from the frame
    def clear_frame(self, target_frame):
        for widgets in target_frame.winfo_children():
            widgets.destroy()

    # method to generate note opbjects and display them in the main window
    def show_notes(self): 
        self.clear_frame(self.frame_notes) # clears any previous display
        self.notes = [] # resets the notes object list
        for note in self.notebook: # create new note objects from the notebook and display them
            new_note = MakeNote(master=self.frame_notes, notebook= self.notebook,note=note)
            self.notes.append(new_note)
            new_note.pack(padx=10, pady=10)
            new_note.config(height= 4, width=50, wraplength=300, justify=tk.LEFT)
        return None
    
    # Method to open an existing notebook file
    def open_notebook(self, default_notebook = None):
        if default_notebook is None:
        # this opens json files
            filepath = filedialog.askopenfilename(initialdir="C:\\Users\\sdemp\\Documents\\GitHub\\Courses\\INST326\\test_files",
                                         filetypes=[("json files", "*.json"), 
                                                    ("csv files", ".csv"),
                                         ("all files", "*.*")])
        else:
            filepath = default_notebook # will open default notebook if none, so every time the program is none
        if filepath:
        # Open the file and load the notebook data
            file = open(filepath, "r")
            self.notebook = []
            dict_list = json.load(file)
            for dict in dict_list:
                note = notesModule.Note(dict)
                self.notebook.append(note)
            self.show_notes() # once the file is loaded, call the method to display the notes in the window
            self.title(filepath)
        else: 
            #error handeling if a filepath is not chosen by user
            self.show_message("Error: You did not select a file so no notebook was loaded")
        return None
    
    # method to show a message box if there is an error on the user's end
    def show_message(self,message):
    # Show an informational message box
        messagebox.showinfo("Alert",message)
    
    # method to save the notebook to a file and refresh the note display
    def save_notebook(self, notebook): 
        self.notebook = notebook
        if self.notebook is not None and len(self.notebook) != 0:
            # Open file dialog to save the notebook to a file
            print('save notebook')
            print(self.notebook)
            # the following code saves the notebook as a json file
            file = filedialog.asksaveasfile(initialdir="C:\\Users\\sdemp\\Documents\\GitHub\\Courses\\INST326\\test_files",
                                              defaultextension=".json", 
                                        title="notebook01",
                                              filetypes=[("json file", ".json"),
                                             ("all files", ".*")])
            if file is not None:
            # Convert notes to dictionaries and save to JSON file
                library = []
                for note in self.notebook:
                    library.append(note.to_dictionary())

                json_out = json.dumps(library, indent=2)
                file.write(json_out)
                file.close()# Close the file after writing the data
                self.show_notes() # this refreshes the notes display in the main window
            else:
                # Close the file after writing the data
                self.show_message("Error: You did not save the notebook to a file")
        else:
             # Show a message if the notebook is empty
            self.show_message("The notebook is empty, nothing was saved")

    # Method to update the scroll region of the canvas based on inner_frame content
    def update_scroll_region(self, event):
        self.canvas_notes.configure(scrollregion=self.canvas_notes.bbox("all"))
        return None


# The NoteForm class provides a Toplevel window that serves as a note entry form, allowing users to input
# and submit details for a new note. The form includes fields for entering the note's title, text, link, 
# tags, author, and snippet, as well as metadata such as date, time, and time zone.
# The class has an __init__() method for initializing the window and a submit() method for handling 
# the submission of new notes. It may also include additional methods to facilitate form clearing 
# and other tasks, such as calculating metadata for the notes.

class NoteForm(tk.Toplevel):
    
    def __init__(self, master, notebook, notes): # initialize the new object
        super().__init__(master) # initialize it as a toplevel window
        # set the new window's default parameters
        self.geometry("800x1000") # set window size
        self.title('New Note') # set window title
        self.notebook = notebook # Store reference to the notebook
        
        # create a frame in the new window that covers the entire window
        self.frame_main = tk.Frame(self)
        self.frame_main.pack(fill=tk.BOTH, expand=True) # Expand frame to fill the window
        self.frame_main.config(bg='light gray') # Set background color of the frame

                
        # Calculate note_id for the new note and set a default note
        self.last_id = len(self.notebook) # Calculate the last note ID
        default_note = notesModule.Note(note_dict ={"title":"new note title",
                    "text":"Lorem ipsum dolor sit amet, consectetur adipiscing elit. Etiam sit amet suscipit mi, non porttitor mauris. Aliquam in lorem risus. Proin mauris mauris, varius ac vulputate sed, tempor nec lacus. Morbi sodales turpis in placerat semper. Donec bibendum blandit ante sit amet hendrerit.", 
                    "link":"If there is a link with this note enter it here.",
                    "tags":"enter hashtags here",
                    "meta":"metadata added at submission","note_id":self.last_id, "author":"name of author","snippet":"print(\"hello world\")"})
        note = default_note
        # Define the notebook as the notebook passed from the main window
        self.notebook = notebook
    
        # Create labels and input fields for the note form using grid layout

        # Title label 
        title_label = tk.Label(self.frame_main, bg='light gray', text='Note Title:')
        title_label.grid(padx=10, pady=10, row=1, column=0, sticky='e')

        # Text label
        text_label = tk.Label(self.frame_main, bg='light gray', text='Note Text:')
        text_label.grid(padx=10, pady=10, row=2, column=0, sticky='e')

        # Link label 
        link_label = tk.Label(self.frame_main, bg='light gray', text='Note Link:')
        link_label.grid(padx=10, pady=10, row=3, column=0, sticky='e')

        # Tags label 
        tag_label = tk.Label(self.frame_main, bg='light gray', text='Note Tags:')
        tag_label.grid(padx=10, pady=10, row=4, column=0, sticky='e')

        # Note ID label
        tag_label = tk.Label(self.frame_main, bg='light gray', text='Note ID:')
        tag_label.grid(padx=10, pady=10, row=5, column=0, sticky='e')

        # Author label 
        tag_label = tk.Label(self.frame_main, bg='light gray', text='Note Author:')
        tag_label.grid(padx=10, pady=10, row=6, column=0, sticky='e')

        # Snippet label 
        tag_label = tk.Label(self.frame_main, bg='light gray', text='Snippet:')
        tag_label.grid(padx=10, pady=10, row=7, column=0, sticky='e')

        # create our note title entry field
        self.note_title = tk.Entry(self.frame_main, width=80)
        self.note_title.grid(padx=10, pady=10, row=1, column=1, sticky='w')
        self.note_title.insert(0, note.title) # adds default text (useful during development)

        # create our note text field
        self.note_text = tk.Text(self.frame_main, height=10, width=60)
        self.note_text.grid(padx=10, pady=10, row=2, column=1)
        self.note_text.insert('1.0', note.text) # adds default text (useful during development)

        # create our note link entry field
        self.note_link = tk.Entry(self.frame_main, width=80)
        self.note_link.grid(padx=10, pady=10, row=3, column=1, sticky='w')
        self.note_link.insert(0, note.link) # adds default text (useful during development)

        # create our note tags entry field
        self.note_tags = tk.Entry(self.frame_main, width=80)
        self.note_tags.grid(padx=10, pady=10, row=4, column=1, sticky='w')
        self.note_tags.insert(0, note.tags) # adds default text (useful during development)

        # create our note id entry field
        self.note_id = tk.Entry(self.frame_main, width=80)
        self.note_id.grid(padx=10, pady=10, row=5, column=1, sticky='w')
        self.note_id.insert(0, note.note_id) # adds default text (useful during development)

        # create our note author entry field
        self.note_author = tk.Entry(self.frame_main, width=80)
        self.note_author.grid(padx=10, pady=10, row=6, column=1, sticky='w')
        self.note_author.insert(0, note.author) # adds default text (useful during development)

        # create our note snippet field
        self.snippet = tk.Text(self.frame_main, height=10, width=60)
        self.snippet.grid(padx=10, pady=10, row=7, column=1)
        self.snippet.insert('1.0', note.snippet) # adds default text (useful during development)        
        
        # Define buttons and place them in the main frame
        # Submit button to submit the note
        b1 = tk.Button(self.frame_main, text='submit', command=self.submit)
        b1.grid(padx=10, pady=10, row=9, column=1, sticky='w')
        # close button to close the note
        b5 = tk.Button(self.frame_main, text='close', command=self.destroy)
        b5.grid(padx=10, pady=10, row=9, column=0) 
        # clear form to clear the default text in note
        b6 = tk.Button(self.frame_main, text='clear form', command=self.clear_form)
        b6.grid(padx=10, pady=10, row=9, column=5) 

    # Method to clear the form fields
    def clear_form(self):
        self.note_title.delete(0,tk.END)
        self.note_text.delete(1.0,tk.END)
        self.note_link.delete(0,tk.END)
        self.note_tags.delete(0,tk.END)
        self.note_author.delete(0,tk.END)
        self.snippet.delete(1.0,tk.END)

    # Method to handle note submission
    def submit(self):
        # calculate the date and time information for the meta field
        now = datetime.datetime.now() # gets the current date and time
        local_now = now.astimezone() # shows the local time and the GMT offset
        local_tz = local_now.tzinfo # Get the local time zone
        created = datetime.datetime.now()
        
        # get all the input values and put them into a dictionary along with the metadata
        title = self.note_title.get()
        text = self.note_text.get('1.0', 'end').strip('\n')
        link = self.note_link.get()
        tags = self.note_tags.get()
        note_id = self.note_id.get()
        author = self.note_author.get()
        snippet = self.snippet.get('1.0', 'end').strip('\n')
        meta = f'note_id {note_id} created {created}, {local_tz} by {author}'
        note_dict = {'title':title, 'text':text, 'link':link, 'tags':tags, 'note_id':note_id, 'author':author, 'snippet':snippet, 'meta':meta}
        
        # add the dictionary to the notebook
        # Create a new note and add it to the notebook
        self.notebook.append(notesModule.Note(note_dict))
        # Increment last_id for future notes
        self.last_id = self.last_id + 1
        # Update the note ID field with the new ID
        self.note_id.delete(0, tk.END)
        self.note_id.insert(0, self.last_id + 1)
        
        return None

# The MakeNote class represents a button for a note in the GUI application.
# It is a subclass of tk.Button and is customized to display note details and handle events
# such as displaying note information and opening an edit window.
# Each MakeNote instance retains all the functionality of a tk.Button while providing additional
# methods for handling note-specific interactions and behaviors.

class MakeNote(tk.Button):
    def __init__(self, master=None, notebook = None, note = notesModule.Note()):
        # Initialize the MakeNote instance as a tk.Button 
        
        super().__init__(master)
        # Store references to the notebook and note data
        self.notebook = notebook
        self.note = note
        

        # Configure the note button with the note title and appearance settings
        self.config(bg='light gray',border = 0,relief="flat",font = ("Arial",9, "underline"), text=f"{self.note.title}")
        
        # Bind mouse events
        self.bind("<Enter>", self.on_hover)
        self.bind("<Leave>", self.on_leave)
        self.bind("<Button-1>", self.note_open)

    def save_close(self):
        # Close the note window and save the notebook
        self.note_window.destroy()
        main_window.save_notebook(notebook=self.notebook)

    def on_hover(self, event): # change the background when the cursor hovers over it
        self.config(bg="lightblue")  

    def on_leave(self, event): # change back when not hovering
        self.config(bg="light gray")  # Restore original color
        
    def note_open(self, event): 
        # Handle note opening by creating a new top window
        
        # create a new top window
        self.note_window = tk.Toplevel(main_window, bg="light gray", height=600, width=600)
        self.note_window.title(self.note.title)
        
        # create a frame in the new window that covers the entire window
        self.frame_main = tk.Frame(self.note_window)
        self.frame_main.pack(fill=tk.BOTH, expand=True)
        self.frame_main.config(bg='light gray')
        
        # create labels in the frame
        title_label = tk.Label(self.frame_main, bg='light gray', text='Note Title:')
        title_label.grid(padx=10, pady=10, row=1, column=0, sticky='e')
        title_content = tk.Label(self.frame_main, bg='light gray', text=self.note.title, wraplength=400, justify=tk.LEFT)
        title_content.grid(padx=10, pady=10, row=1, column=1, sticky='w')        

        text_label = tk.Label(self.frame_main, bg='light gray', text='Note Text:')
        text_label.grid(padx=10, pady=10, row=2, column=0, sticky='e')
        text_content = tk.Label(self.frame_main, bg='light gray', text=self.note.text, wraplength=400, justify=tk.LEFT)
        text_content.grid(padx=10, pady=10, row=2, column=1, sticky='w')

        link_label = tk.Label(self.frame_main, bg='light gray', text='Note Link:')
        link_label.grid(padx=10, pady=10, row=3, column=0, sticky='e')
        link_content = tk.Label(self.frame_main, bg='light gray', text=self.note.link, wraplength=400, justify=tk.LEFT)
        link_content.grid(padx=10, pady=10, row=3, column=1, sticky='w')
        
        tag_label = tk.Label(self.frame_main, bg='light gray', text='Note Tags:')
        tag_label.grid(padx=10, pady=10, row=4, column=0, sticky='e')
        tag_content = tk.Label(self.frame_main, bg='light gray', text=self.note.tags, wraplength=400, justify=tk.LEFT)
        tag_content.grid(padx=10, pady=10, row=4, column=1, sticky='w')

        note_id_label = tk.Label(self.frame_main, bg='light gray', text='Note ID:')
        note_id_label.grid(padx=10, pady=10, row=5, column=0, sticky='e')
        note_id_content = tk.Label(self.frame_main, bg='light gray', text=self.note.note_id, wraplength=400, justify=tk.LEFT)
        note_id_content.grid(padx=10, pady=10, row=5, column=1, sticky='w')

        author_label = tk.Label(self.frame_main, bg='light gray', text='Note Author:')
        author_label.grid(padx=10, pady=10, row=6, column=0, sticky='e')
        author_content = tk.Label(self.frame_main, bg='light gray', text=self.note.author, wraplength=400, justify=tk.LEFT)
        author_content.grid(padx=10, pady=10, row=6, column=1, sticky='w')

        snippet_label = tk.Label(self.frame_main, bg='light gray', text='Snippet:')
        snippet_label.grid(padx=10, pady=10, row=7, column=0, sticky='e')
        snippet_content = tk.Label(self.frame_main, bg='light gray', text=self.note.snippet, wraplength=400, justify=tk.LEFT)
        snippet_content.grid(padx=10, pady=10, row=7, column=1, sticky='w')
        
        meta_label = tk.Label(self.frame_main, bg='light gray', text='Note Meta:')
        meta_label.grid(padx=10, pady=10, row=8, column=0, sticky='e')
        meta_content = tk.Label(self.frame_main, bg='light gray', text=self.note.meta, wraplength=400, justify=tk.LEFT)
        meta_content.grid(padx=10, pady=10, row=8, column=1, sticky='w')        

        # create a button to close the note window
        b10 = tk.Button(self.frame_main, text='save and close', command=self.save_close)
        b10.grid(padx=10, pady=10, row=9, column=0)

        # create a button to edit the note
        b11 = tk.Button(self.frame_main, text='edit note', command=self.note_edit)
        b11.grid(padx=10, pady=10, row=9, column=1)

    def save_edits(self):
        # calculate the date and time information for the meta field
        now = datetime.datetime.now() # gets the current date and time
        local_now = now.astimezone() # shows the local time and the GMT offset
        local_tz = local_now.tzinfo 
        edited = datetime.datetime.now()
        
        # get all the input values and put them into a dictionary along with the metadata
        self.note.title = self.title_content.get()
        self.note.text = self.text_content.get('1.0', 'end').strip('\n')
        self.note.link = self.link_content.get()
        self.note.tags = self.tag_content.get()
        self.note.note_id = self.note_id_content.get()
        self.note.author = self.author_content.get()
        self.note.snippet = self.snippet_content.get('1.0', 'end').strip('\n')
        
        # modify the metadata to track edits
        self.note.meta = self.note.meta + f', edited {edited}, {local_tz}'
        note_dict = {'title':self.note.title, 'text':self.note.text, 'link':self.note.link, 'tags':self.note.tags, 'note_id':self.note.note_id, 'author':self.note.author, 'snippet':self.note.snippet, 'meta':self.note.meta}
        new_note = notesModule.Note(note_dict)

        new_notebook = []
        for note in self.notebook:
            if note.note_id != self.note.note_id:
                new_notebook.append(note)
            else:
                new_notebook.append(new_note)
        self.notebook = new_notebook
        
        # destroy the edit window
        self.edit_window.destroy()
        self.note_window.destroy()
        event = None
        self.note_open(event)
        
        
    def note_edit(self):
        # loads note into an editable window
        # create a new top window
        self.edit_window = tk.Toplevel(main_window, bg="light gray", height=700, width=600)
        self.edit_window.title(self.note.title)


        # create a frame in the new window that covers the entire window
        self.frame_main = tk.Frame(self.edit_window)
        self.frame_main.pack(fill=tk.BOTH, expand=True)
        self.frame_main.config(bg='light gray')
        
        # create labels in the frame
        title_label = tk.Label(self.frame_main, bg='light gray', text='Note Title:')
        title_label.grid(padx=10, pady=10, row=1, column=0, sticky='e')
        self.title_content = tk.Entry(self.frame_main, width=80)
        self.title_content.grid(padx=10, pady=10, row=1, column=1, sticky='w')
        self.title_content.insert(0, self.note.title)

        text_label = tk.Label(self.frame_main, bg='light gray', text='Note Text:')
        text_label.grid(padx=10, pady=10, row=2, column=0, sticky='e')
        self.text_content = tk.Text(self.frame_main, height=8, width=60)
        self.text_content.grid(padx=10, pady=10, row=2, column=1, sticky='w')
        self.text_content.insert('1.0', self.note.text)

        link_label = tk.Label(self.frame_main, bg='light gray', text='Note Link:')
        link_label.grid(padx=10, pady=10, row=3, column=0, sticky='e')
        self.link_content = tk.Entry(self.frame_main, width=80)
        self.link_content.grid(padx=10, pady=10, row=3, column=1, sticky='w')
        self.link_content.insert(0, self.note.link)
        
        tag_label = tk.Label(self.frame_main, bg='light gray', text='Note Tags:')
        tag_label.grid(padx=10, pady=10, row=4, column=0, sticky='e')
        self.tag_content = tk.Entry(self.frame_main, width=80)
        self.tag_content.grid(padx=10, pady=10, row=4, column=1, sticky='w')
        self.tag_content.insert(0, self.note.tags)

        note_id_label = tk.Label(self.frame_main, bg='light gray', text='Note ID:')
        note_id_label.grid(padx=10, pady=10, row=5, column=0, sticky='e')
        self.note_id_content = tk.Entry(self.frame_main, width=80)
        self.note_id_content.grid(padx=10, pady=10, row=5, column=1, sticky='w')
        self.note_id_content.insert(0, self.note.note_id)

        author_label = tk.Label(self.frame_main, bg='light gray', text='Note Author:')
        author_label.grid(padx=10, pady=10, row=6, column=0, sticky='e')
        self.author_content = tk.Entry(self.frame_main, width=80)
        self.author_content.grid(padx=10, pady=10, row=6, column=1, sticky='w')
        self.author_content.insert(0, self.note.author)

        snippet_label = tk.Label(self.frame_main, bg='light gray', text='Snippet:')
        snippet_label.grid(padx=10, pady=10, row=7, column=0, sticky='e')
        self.snippet_content = tk.Text(self.frame_main, height=8, width=60)
        self.snippet_content.grid(padx=10, pady=10, row=7, column=1, sticky='w')
        self.snippet_content.insert('1.0', self.note.snippet)
        
        meta_label = tk.Label(self.frame_main, bg='light gray', text='Note Meta:')
        meta_label.grid(padx=10, pady=10, row=8, column=0, sticky='e')
        self.meta_content = tk.Label(self.frame_main, bg='light gray', text=self.note.meta, wraplength=400, justify=tk.LEFT)
        self.meta_content.grid(padx=10, pady=10, row=8, column=1, sticky='w')        

        # create a button to close the edit window
        b10 = tk.Button(self.frame_main, text='close', command=self.edit_window.destroy)
        b10.grid(padx=10, pady=10, row=9, column=0)

        # create a button to edit the note
        b11 = tk.Button(self.frame_main, text='save edits', command=self.save_edits)
        b11.grid(padx=10, pady=10, row=9, column=1)
        
        
        
# main execution
if __name__ == '__main__':
    
    main_window = MainWindow() # this creates a notebook / main window called main_window. You may change the name if you want
    try:
        main_window.open_notebook("JillianNotebook.json") # Make sure it is in the same folder as the python file 
    except FileNotFoundError: # error handeling if the notebook does not exist 
        main_window.show_message("default notebook was not found")

    main_window.mainloop()
```
